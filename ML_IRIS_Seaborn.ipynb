```python
import seaborn as sns; sns.set(style="ticks", color_codes=True)
iris = sns.load_dataset("iris")
plots1 = sns.pairplot(iris)
print(plots1)
```

```python
plots2 = sns.pairplot(iris, hue="species");
print(plots2)
```

```python
plots_3 = sns.pairplot(iris, hue="species", markers=["o", "s", "D"]);
print(plots_3)
```

```python
plot_subset_var = sns.pairplot(iris, vars=["sepal_width", "sepal_length"]);
print(plot_subset_var)
```

```python
plot_with_density_estimate = sns.pairplot(iris, diag_kind="kde");
print(plot_with_density_estimate)
```

```python
plot_with_reg = sns.pairplot(iris, kind="reg");
print(plot_with_reg)
```
